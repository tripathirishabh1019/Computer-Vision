```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np # Import numpy

# Load the image in color mode (BGR is the default)
img = cv2.imread('/content/Mountains.jpg',1)

# Check if the image was loaded successfully
if img is not None:
    # Split the image into its BGR channels
    imgBlue = img[:,:,0]
    imgGreen = img[:,:,1]
    imgRed = img[:,:,2]

    # Stack the channels horizontally
    new_img = np.hstack((imgBlue, imgGreen, imgRed))

    # Display the new image using cv2_imshow (suitable for Colab)
    cv2_imshow(new_img)
    # cv2.waitKey(0) # Not needed with cv2_imshow in Colab
    # cv2.destroyAllWindows() # Not needed with cv2_imshow in Colab
else:
    print("Error loading image.")
```
