```python
import cv2
import numpy as np
from IPython.display import display, Image
import urllib.request

# Download a pre-trained Haar Cascade classifier for face detection
# You can find more classifiers in the OpenCV data directory on GitHub
# Corrected URL for the Haar Cascade XML file
face_cascade_url = 'https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml'
face_cascade_filename = 'haarcascade_frontalface_default.xml'

# Download the file if it doesn't exist
try:
    urllib.request.urlretrieve(face_cascade_url, face_cascade_filename)
except Exception as e:
    print(f"Error downloading the cascade file: {e}")
    # Handle the error, perhaps exit or use a locally available file

# Load the cascade
face_cascade = cv2.CascadeClassifier(face_cascade_filename)

# Check if the cascade classifier loaded successfully
if face_cascade.empty():
    print("Error loading cascade classifier.")
else:
    # Load an image to perform face detection on
    # Replace '/content/Face.jpg' with the path to your actual image file with faces
    img_path = '/content/Face.jpg' # Make sure this path points to an image file
    img = cv2.imread(img_path)

    # Check if image is loaded
    if img is not None:
        # Convert to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Detect faces in the image
        # Parameters: scaleFactor, minNeighbors
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)

        # Draw a rectangle around the faces
        for (x, y, w, h) in faces:
            cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

        # Display the output image
        is_success, im_buf = cv2.imencode(".png", img)
        if is_success:
            display(Image(data=im_buf.tobytes()))
        else:
            print("Error encoding image for display.")

    else:
        print(f"Error: Could not load image from {img_path}. Please check the file path.")
```
